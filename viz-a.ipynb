```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


sns.set()
sns.get_dataset_names()
```

```python

df = sns.load_dataset('titanic')
df.head()
```

## Bar Plot for first 10 record

```python
sns.barplot(x='pclass', y='fare', data=df.head(10))
```

## Relational Plot

```python
sns.relplot(x='pclass', y='fare', data=df)
```

## summary

the pclass increase with fare
